```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
sns.set(rc={'figure.figsize':(8,6)})
```

```python
def euc_distance(a,b):
    return np.sqrt((a[0]- b[0])**2 + (a[1]-b[1])**2)
```

```python
def f(x):
    return ((5*x[0]-x[1])**4 + (x[0]-2)**2 + x[0] - 2*x[1] + 12)
```

```python
def exact_line(f,d,x):
    min_val = 1000000
    best_alpha = 0
    interval = np.arange(-100,100,0.005)
    for alpha in interval:
        if(f(x+(alpha*np.array(d)))<= min_val):
            min_val = f(x+(alpha*np.array(d)))
            best_alpha = alpha
    return best_alpha
```

```python
alpha_vals = []
x_vals = []
x_next_vals = []
directions = []
def cyclic_search(eps, f):
    x = np.zeros((100,2))
    k = 0
    x[0] = [5,35]
    I = np.identity(2)
    while(True):
        for j in range(2):
            alpha = 0.
            alpha = exact_line(f,I[j],x[k])
            alpha_vals.append(alpha)
            directions.append(I[j])
            x[k+1][j] = x[k][j] + alpha
            x[k+1][1-j] = x[k][1-j]
            x_vals.append(x[k])
            x_next_vals.append([round(i,4) for i in x[k+1]])
            k= k+1
        if(euc_distance(x[k],x[k-1]) <= eps):
            break
    return x[k]
```

```python
output = cyclic_search(0.0001, f)
df_cyclic = pd.DataFrame(x_vals, columns = ["X(k)" , "Y(k)"])
df_cyclic["f(X(k))"] = [f(x) for x in x_vals]
df_cyclic["d(k)"] = directions
df_cyclic["alpha(k)"] = alpha_vals
df_cyclic["X(k+1)"] = x_next_vals
```

```python
print("Results : ")
df_cyclic
```

```python
print(f"""

x* =  {output}

f(x*) = {f(output)}

""")
```

# Hook-Jeeves Method

## Parameter Set 1

For exact line search (a,b,eps): (-100,100,0.005)

For Hook-Jeeves Method (eps) : (0.001)

Initial point : [10,10]

```python
def exact_line(f,d,x):
    min_val = 1000000
    best_alpha = 0
    interval = np.arange(-100,100,0.005)
    for alpha in interval:
        if(f(x+(alpha*np.array(d)))<= min_val):
            min_val = f(x+(alpha*np.array(d)))
            best_alpha = alpha
    return best_alpha
```

```python
alpha_vals = []
x_vals = []
x_next_vals = []
x_temp_vals = []
directions = []
def hook_jeeves_method(eps, f):
    x = np.zeros((500,2))
    d = np.zeros((500,2))
    Lambda = np.zeros(500)
    k = 0
    x[0] = [10,10]
    I = np.identity(2)
    while(True):
        alphas = []
        x_temp = [[0.],[0.]]
        for j in range(2):
            alpha = 0.
            alpha = exact_line(f,I[j],x[k])
            alphas.append(alpha)
            x_temp[j] = x[k][j] + alpha
        alpha_vals.append(alphas)
        x_temp_vals.append(np.round(x_temp, 4))
        d[k] = x_temp - x[k]
        directions.append(np.round(d[k],4))
        Lambda[k] = exact_line(f,d[k],x[k])
        x[k+1] = x[k] +  Lambda[k] * d[k]
        x_next_vals.append([round(i,4) for i in x[k+1]])
        x_vals.append(x[k])
        k= k+1
        if(euc_distance(x[k],x[k-1]) <= eps):
            break
    return x[k]
```

```python
output = hook_jeeves_method(0.001, f)
df_hook = pd.DataFrame(x_vals, columns = ["X(k)" , "Y(k)"])
df_hook["f(X(k))"] = [f(x) for x in x_vals]
df_hook["X_tempk(k)"] = x_temp_vals
df_hook["d(k)"] = directions
df_hook["alpha(k)"] = alpha_vals
df_hook["X(k+1)"] = x_next_vals
```

```python
print("Results : ")
df_hook
```

```python
print(f"""

x* =  {output}

f(x*) = {f(output)}

""")
```

# Simplex Search

```python
def simplex_search(alpha ,beta ,gamma, eps, f , start_pts):
    
    x = np.zeros((3,2))
    x_h = np.array(start_pts[np.argmax([f(i) for i in start_pts])])
    h = np.argmax([f(i) for i in start_pts])
    x_l = np.array(start_pts[np.argmin([f(i) for i in start_pts])])
    l = np.argmin([f(i) for i in start_pts])
    for i in range(0,len(start_pts)):
        if((i != np.argmax([f(i) for i in x])) or (i !=np.argmin([f(i) for i in x]))):
            x_m = start_pts[i]
            m = i-1

    x = [start_pts[l],start_pts[m],start_pts[h]]
    
    
    while(True):
        x_h = np.array(x[np.argmax([f(i) for i in x])])
        h = np.argmax([f(i) for i in x])
        x_l = np.array(x[np.argmin([f(i) for i in x])])
        l = np.argmin([f(i) for i in x])
                      
        for i in range(0,len(start_pts)):
            if((i != np.argmax([f(i) for i in x])) or (i !=np.argmin([f(i) for i in x]))):
                x_m = x[i]
                m = i-1
            
        counter = 0
        Sum = np.zeros(2)
        
        for i in range(0,len(x)):
            if(np.array_equal(x[i], x_h) == False):
                counter +=1
                Sum += x[i]
        
        x_bar = Sum/counter
        x_bar = np.array(x_bar)
        
        # Reflection
        x_r = np.array(x_bar) + alpha * (x_bar-x_h)
        x_r = np.array(x_r)
        
        if(f(x_l)>f(x_r)):
            
            #expansion
            x_e = x_bar + gamma * (x_r-x_bar)
            x_e = np.array(x_e)
            if(f(x_r)>f(x_e)):
                x_h = x_e
                x[h] = x_h
            else:
                x_h = x_r #expansion failed
                x[h] = x_h
        else:
            values = []
            for i in range(0,len(x)):
                
                if(np.array_equal(x[i], x_h) == False):
                    values.append(f(x[i]))
                    
            Max = np.max(values)
            
            if(Max >= f(x_r)):
                x_h = x_r
                x[h] = x_h
            
            else:
                
                #contraction
                if(f(x_h) > f(x_r)):
                    x_h_prime = x_r
                    x_h_prime = np.array(x_h_prime)
                else:
                    x_h_prime = x_h
                    x_h_prime = np.array(x_h_prime)
                x_c = x_bar + beta * (x_h_prime-x_bar)
                
                if(f(x_c)<= f(x_h_prime)):
                    x_h = x_c
                    x[h] = x_h
                else:
                    for i in range(0,len(x)):
                        x[i] = x[i] +  (x_l-x[i]) / 2
        
        if(np.sqrt(1/(len(x) + 1) * np.sum([(f(i) - f(x_bar))**2 for i in x])) < eps):
            break
        
        
        print(x_bar)
    return x_bar
```
